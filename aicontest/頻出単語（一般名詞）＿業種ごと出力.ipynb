```python
import MeCab
import sys
import re
from collections import Counter
import glob
import numpy as np
import csv
import pprint
import os
import pandas as pd
```

```python
# 対象のファイルパスの入力
filepath = './input/33/*.txt'
```

```python
# Taggerの定義
tagger= MeCab.Tagger()
# text格納用リスト型変数
file_text = [] 
# フォルダ名取得用変数
subdirname = os.path.basename(os.path.dirname(filepath))
#muliti files
for file in glob.glob(filepath): #Pythonソースと同じディレクトリの「test」フォルダにあるテキストファイルが対象
    #file_name.append(os.path.basename(file)) #ファイルパスからファイル名のみ抽出
    file_data = open(file,'r',encoding='utf-8-sig') 
    bindata = file_data.read()
    parse_data = tagger.parse(bindata)
    lines = parse_data.split('\n')
    items = (re.split('[\t,]', line) for line in lines)
    # 名詞をリストに格納
    words = [item[0]
             for item in items
             if (item[0] not in ('EOS', '', 't', 'ー') and
                 item[1] == '名詞' and item[2] == '一般')]
    file_text.extend(words) 

# 頻度順に出力
# Ref: https://note.nkmk.me/python-collections-counter/
counter = Counter(file_text)
word_file = []
num = len(file_text)
num_file = []
ratio_file = []
#print("出現単語数：" + str(num))
for word, count in counter.most_common():
    word_file.append(word)
    num_file.append(str(count))
    ratio = round (count*100/num,3)
    ratio_file.append(str(ratio)+"%")
    #print(f"{word}: {count}")

# Pandasでデータ作成
df = pd.DataFrame({'単語': word_file, '出現数': num_file, f'割合（出現数/総数:{num}）': ratio_file})

# CSVファイルを作成
df.to_csv(f'./output/頻出単語出力/出現単語リスト_業種_{subdirname}.csv', sep=',', encoding='utf_8_sig')
pd.read_csv(f'./output/頻出単語出力/出現単語リスト_業種_{subdirname}.csv', sep = ':', encoding='utf_8_sig')
```
